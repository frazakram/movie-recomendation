```python
def convert(text):
    L = []
    for i in ast.literal_eval(text):
        L.append(i['name']) 
    return L 
```

```python
import ast
ast.literal_eval('[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]')
```

```python

def convert3(text):
    L = []
    counter = 0
    for i in ast.literal_eval(text):
        if counter < 3:
            L.append(i['name'])
        counter+=1
    return L 
```

```python
moviess1.head(4)
```

```python
moviess1['cast'] = moviess1['cast'].apply(convert3)
```

```python
moviess1.head(3)
```

```python
def fetch_director(text):
    L = []
    for i in ast.literal_eval(text):
        if i['job'] == 'Director':
            L.append(i['name'])
    return L 
```

```python
moviess1['crew'] = moviess1['crew'].apply(fetch_director)
```

```python
moviess1['cast']=moviess1['cast'].apply(lambda x:[i.replace(" ","")for i in x])
moviess1['crew']=moviess1['crew'].apply(lambda x:[i.replace(" ","")for i in x])
moviess1['keywords']=moviess1['keywords'].apply(lambda x:[i.replace(" ","")for i in x])
moviess1['genres']=moviess1['genres'].apply(lambda x:[i.replace(" ","")for i in x])
```

```python
moviess1['tags'] = moviess1['overview'] + moviess1['genres'] + moviess1['keywords'] + moviess1['cast'] + moviess1['crew']
```

```python
newmovies= moviess1.drop(columns=['overview','genres','keywords','cast','crew'])
```

```python
newmovies.head(4)
```

```python
newmovies['tags'].apply(lambda x:" ".join(x))
```

```python
def try_join(l):
    try:
        return ' '.join(map(str, l))
    except TypeError:
        return np.nan

newmovies['tags'] = [try_join(l) for l in newmovies['tags']]
```

```python
newmovies.head(3)
```

```python
newmovies['tags'][0]
```

```python
#  vectorization
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv=CountVectorizer(max_features=5000,stop_words='english')
```

```python
vector=cv.fit_transform(newmovies['tags']).toarray()
```

```python
vector.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vector)
```

```python
similarity
```

```python
newmovies[newmovies['title'] == 'The Lego Movie'].index[0]
```

```python
def recommend(movie):
    index = newmovies[newmovies['title'] == movie].index[0]
    distances = sorted(list(enumerate(similarity[index])),reverse=True,key = lambda x: x[1])
    for i in distances[1:6]:
        print(newmovies.iloc[i[0]].title)
        
```

```python
recommend('Gandhi')
```

```python
recommend('Avatar')
```
